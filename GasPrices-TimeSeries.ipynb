```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
gasPrices = pd.read_csv('GasPrices.csv')
```

```python
gasPrices.head()
```

```python
gasPrices['YYYYMM'] = pd.to_datetime(gasPrices['YYYYMM'], format='%Y%m').dropna()
```

```python
gasPrices.head()
```

```python
graph = gasPrices.plot(
    x = 'YYYYMM',
    y = 'Value',
    color = ['#C72C41'],
    figsize = (8,5)
)

plt.suptitle(" Unleaded Regular Gasoline, U.S. City Average Retail Price (1976-2023) ", fontsize=11, fontweight='bold')
plt.title(" U.S. Energy Information Administration ", fontsize=10)

labels = ['Price']
graph.legend(labels)
graph.set_xlabel('Year')
graph.set_ylabel('Price')
graph.grid(linewidth = 1.5, alpha = 0.20)
plt.show()
```
